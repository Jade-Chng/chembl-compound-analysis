```python
#Imports
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('Chemical.csv', delimiter=';')
```

```python
df
```

```python
act = pd.read_csv('Activity_1.csv', delimiter=';')
```

```python
act
```

```python
#Merge
mydf = pd.merge(df, act, on='ChEMBL ID')
```

```python
#Standardise the IC50 values
mydf['Standard'] = np.where(mydf['IC50 unit'] == 'nM', mydf['IC50 value']*1e-9, mydf['IC50 value']*1e-6)
```

```python
mydf
```

```python
#pIC50
mydf['pIC50'] = -np.log10(mydf['Standard'])
```

```python
mydf
```

##Data Analysis
1. Find the mean pIC50
2. Find the name of the most active compound.
3. List the number of compounds whose molecular weight range between 200 and 350 g/mol (inclusive of 200 and 350 g/mol)?

```python
#1. Find the mean pIC50
np.mean(mydf['pIC50'])
print(f"The mean pIC50 value is {np.mean(mydf['pIC50'])}")
```

```python
#2. Find the name of the most active compound.
#Step 1: Find highest pIC50 value
maxpic50 = mydf['pIC50'].idxmax()

#Step 2: Find its name
mydf.loc[maxpic50, 'Name']

#Put it in full sentenece
print(f"The most active compound is {mydf.loc[maxpic50, 'Name']}")
```

```python
#Amount of compounds whose between 200 and 350 in MW
((mydf['Molecular Weight'] > 200) & (mydf['Molecular Weight'] < 350)).sum()
```

```python
totalCpd = ((mydf['Molecular Weight'] > 200) & (mydf['Molecular Weight'] < 350)).sum()
print(f"There are {totalCpd} compounds whose molecular weight range between 200 and 350 g/mol (inclusive of 200 and 350 g/mol)")
```

## Visualizing Data
Python supports the matplotlib library; which provides extensive graphical capabilities.  Additionally, the Python Pandas library and the Seaborn library add a higher level graphics capability. Pandas and Seaborn abstract a lot of the low level details. As you will see, since these libraries are based on Matplotlib, you can always add needed details. These features make Python a useful language to create visualizations of your data when exploring relationships between the data features.  Further, you can identify features that may be useful for predicting labels in machine learning projects.


```python
#Bar plot
sns.catplot(data=mydf,
            x="#RO5 Violations",
            y = "Molecular Weight",
            kind='bar').set(title='Molecular Weight of different RO5 Violations groups')
```

```python
#Strip plot
sns.catplot(data = mydf,
            x = "#RO5 Violations",
            y = "Molecular Weight",
            kind = "strip"
            ).set(title = 'Molecular Weight of different RO5 Violations groups')
```

```python
#Swarm plot
sns.catplot(data = mydf,
            x = "#RO5 Violations",
            y = "Molecular Weight",
            kind = "swarm"
            ).set(title = 'Molecular Weight of different RO5 Violations groups')
```

```python
#Boxplot
sns.catplot(data = mydf,
            x = "#RO5 Violations",
            y = "Molecular Weight",
            kind = "box"
            ).set(title = 'Molecular Weight of different RO5 Violations groups')
```

```python
#Boxen plot
sns.catplot(data = mydf,
            x = "#RO5 Violations",
            y = "Molecular Weight",
            hue = "#RO5 Violations",
            palette = 'husl',
            kind = "boxen"
            ).set(title = 'Molecular Weight of different RO5 Violations groups')
```

Plot the histogram to visualise the molecular weight of your dataset. Add two different lines to represent the mean and median respectively. Add the annotation of the line in the legend.

```python
#Add mean and median line in the histogram
fig, ax = plt.subplots()
sns.histplot(mydf['Molecular Weight'], ax=ax)

#Assign mean value
meanMW = np.mean(mydf['Molecular Weight'])

#Assign median value
medianMW = np.median(mydf['Molecular Weight'])

#Add lines
ax.axvline (x = meanMW, color = 'b', label = 'Mean', linestyle = '-', linewidth = 2)
ax.axvline (x = medianMW, color = 'b', label = 'Median', linestyle = '--', linewidth = 2)

#Show legend
ax.legend()
```
